## EDA and Data Analysis

Main Focus:

- Visualizing class imbalance
- Comparing feature distributions across classes
- Spotting signal vs noise

### Class Imbalance Visualization

```python
import pandas as pd
```

```python
df_cleaned = pd.read_csv('../data/processed/student_mat_cleaned.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='at_risk', data=df_cleaned)
plt.title("Class InBalance: At Risk vs On Track")
```

### Box Plot for Key Features

```python
features = ["studytime", "absences", "goout", "Dalc", "Walc", "freetime", "health"]

for col in features:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='at_risk', y=col, data=df_cleaned)
    plt.title(f"Box Plot of {col} by At Risk Status")
    plt.show()
```
